```python
from pymongo import MongoClient
import pandas as pd
client = MongoClient()
client = MongoClient(host="localhost", port=27017)
```

```python
db = client['Amusement_Park']
```

## Query - 1
### Find all activities in the rides collection

```python
collection1 = db["RIDES"]
result1 = collection1.find({})
df1 = pd.DataFrame(result1)
df1
```

## Query 2
### Find customers who provided feedback with ratings greater than 4.

```python
collection2 = db["FEEDBACK"]
query = {"Ratings": {"$gt": 4}}
result = collection2.find(query)
df2 = pd.DataFrame(result)
df2
```

## Query 3
### Calculate average ratings per activity.

```python
map_reduce_result = db.command({
    'mapReduce': 'FEEDBACK',
    'map': 'function() { emit(this.Activity_ID, this.Ratings); }',
    'reduce': 'function(key, value) { return Array.sum(value) / value.length; }',
    'out': {'inline': 1},
    'finalize': 'function(key, reducedValue) { var roundedValue = Math.round(reducedValue * 100) / 100; return roundedValue ; }'
})

df3 = pd.DataFrame(map_reduce_result['results'])
df3.rename(columns={'value': 'Average_Ratings'}, inplace=True)
df3
```
